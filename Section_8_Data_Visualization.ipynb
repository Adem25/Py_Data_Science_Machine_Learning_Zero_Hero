```python
x_axis = [1,2,3]
y_axis = [4,5,6]

plt.title('My First Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x_axis, y_axis, 'red')
```

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]

x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]

plt.title('Two Plots on One Chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.plot(x1_axis, y1_axis, 'r')
plt.plot(x2_axis, y2_axis, 'g')
plt.legend(['exponential', 'linear'])
```

```python
values = [10,15,20,10,15]
plt.hist(values) # used to represent the frquency distribution of continuous variables. x values represent ranges of data
```

```python
plt.hist(values, color='r', width=0.5)
```

```python
values_bar = [10,15,20]
# is a diagrammic comparison of discrete variables. x values are individual entities.
# Secound argument is the height of each bar
plt.bar(values_bar, [0.5,1,2], width=2)
```

```python
values_pie = [20,20,35,25]
plt.pie(values_pie, labels=['a','b','c','d'], explode=[0,0,0.2,0], colors=['r','g','b', 'y'])
```

```python
x1_axis = [2,4,6,8]
y1_axis = [1,10,100,1000]

x2_axis = [1,3,5,7]
y2_axis = [100,110,120,130]

plt.subplot(1,2,1)
plt.title('Exponential')
plt.plot(x1_axis, y1_axis,'r')

plt.subplot(1,2,2)
plt.title('Linear')
plt.plot(x2_axis, y2_axis,'g')

plt.tight_layout()
```
