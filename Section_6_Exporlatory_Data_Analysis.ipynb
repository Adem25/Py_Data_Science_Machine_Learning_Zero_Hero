```python
dfTitanic = pd.read_csv('titanic.csv')
dfTitanic
```

```python
sns.scatterplot(x=dfTitanic['Fare'], y=dfTitanic['Age'])
plt.title('Bivariate Analysis')
```

```python
Image("correlation_plot.png")
```

```python
# Korrelation bedeutet, dass zwei Dinge miteinander zusammenhängen oder in Beziehung zueinander stehen. 
# Wenn zwei Variablen korrelieren, bedeutet dies, dass eine Veränderung in einer Variable oft 
# mit einer Veränderung in der anderen Variable einhergeht. Wenn beispielsweise die Temperatur steigt, 
# steigt auch die Anzahl der verkauften Eiscremebecher. 
# In diesem Fall besteht eine positive Korrelation zwischen Temperatur und Eiscremeverkäufen. 
# In einfachen Worten ausgedrückt, lässt sich sagen: "Wenn eine Sache passiert, passiert auch oft die andere."
dfTitanic[['Fare', 'Age']].corr()
```

```python
# Detecting outliers - remove with Z-Score
x = pd.Series([1,2,3,6,7,8,150])
mean = x.mean()
std = x.std()
z_scores = abs((x-mean)/std)
z_scores
```

```python
outliers_removed = x[z_scores <= 1.5]
outliers_removed
```

```python
# Detecting outliers - replace with Median
x = pd.Series([1,2,3,6,7,8,150])
mean = x.mean()
std = x.std()
median = np.median(x)
z_scores = abs((x-mean)/std)
median
```

```python
x[z_scores > 1.5] = median
x
```
